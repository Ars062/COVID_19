```python
import pandas as pd
import numpy as np
import datetime as dt
import seaborn as sns
import matplotlib.pyplot as plt
import sklearn
```

```python
Covid=pd.read_csv("D:\python_da\covid_19\covid_19_dataset.csv",parse_dates=['Date'],dayfirst=True)
```

```python
Covid.tail(100)
```

```python
Covid=Covid[['Date','Province/State','Country/Region','Time','Confirmed','Deaths','Recovered']]
Covid.columns=['Date','State','Country','Time','Confirmed','Deaths','Recovered']
```

```python
Covid.tail(18000)
```

```python
today=Covid[Covid.Date == '01/22/2020']
```

```python
today.head()
```

```python
#Max_Confirmed cases in any day by states
max_Confirmed_cases=today.sort_values(by='Confirmed',ascending=False)
max_Confirmed_cases[0:5]
```

```python
sns.set(rc={'figure.figsize':(15,20)})
sns.barplot(x="State",y="Confirmed",data=max_Confirmed_cases[0:5],hue="State")
plt.show
```

```python
# sorting w.r.t to deaths in any day
Max_death =today.sort_values(by='Deaths',ascending=False)
sns.set(rc={'figure.figsize':(10,10)})
sns.barplot(x="State",y="Deaths",hue="State",data=Max_death[0:5])
```

```python
# sorting w.r.t to recovered in any day
Max_Recovered=today.sort_values(by="Recovered",ascending=False)
sns.set(rc={'figure.figsize':(10,10)})
sns.barplot(x="State",y="Recovered",hue="State",data=Max_Recovered[0:5])
```
